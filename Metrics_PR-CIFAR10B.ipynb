```python
import pickle

pkl_latest = './metrics/latest/060221-234441_CIFAR10_pr.pkl'
pkl_fewshot = './metrics/fewshot/060321-015613_CIFAR10_pr.pkl'

with open(pkl_latest, 'rb') as f:
    pr_latest = pickle.load(f)

with open(pkl_fewshot, 'rb') as f:
    pr_fewshot = pickle.load(f)
```

```python
import pandas as pd

CLASS = 0

data = []

sampler_dict = {
    'grad_based_mean': 'GradSamplerMean',
    'grad_based_std': 'GradSamplerVar',
    'rand': 'RandSampler',
    'kmeans': 'KMeansSampler',
    'loss_based_mean': 'LossSamplerMean',
    'loss_based_std': 'LossSamplerVar'
}
for pr in pr_fewshot:
    # ds is CIFAR10, so no need to grab
    sampler = pr['sampler_type']
    num_examples = pr['num_examples']
    ap = pr['lr_byol_pr']['average_precision'][CLASS]
    
    _dict = {
        'sampler': sampler_dict[sampler],
        'num_examples': num_examples,
        'ap': ap
    }    
    data.append(_dict)
    
data = pd.DataFrame(data=data)
    
data = data.groupby(["sampler", "num_examples"]).mean().reset_index()
res = data.sort_values(['num_examples', 'sampler'], ascending=(True, False))
res = res.round(3)
print(res.to_latex(index=False))
```
